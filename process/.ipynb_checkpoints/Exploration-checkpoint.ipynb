```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import datetime
%matplotlib inline

cs = ['#F0F8FF', '#FAEBD7', '#00FFFF', '#7FFFD4', '#F0FFFF', '#F5F5DC', '#FFE4C4', '#000000', '#FFEBCD', '#0000FF', '#8A2BE2', '#A52A2A', '#DEB887', '#5F9EA0', '#7FFF00', '#D2691E', '#FF7F50', '#6495ED', '#FFF8DC', '#DC143C', '#00FFFF', '#00008B', '#008B8B', '#B8860B', '#A9A9A9', '#006400', '#BDB76B', '#8B008B', '#556B2F', '#FF8C00', '#9932CC', '#8B0000', '#E9967A', '#8FBC8F', '#483D8B', '#2F4F4F', '#00CED1', '#9400D3', '#FF1493', '#00BFFF', '#696969', '#1E90FF', '#FF0000']

races = pd.read_csv('../dataset/races.csv')
gps = { gp: ((i + 1) * 10) + 4 for i, gp in enumerate(dict.fromkeys(races['name'])) }
gps_colors = {gp: cs[i] for i, gp in enumerate(dict.fromkeys(races['name'])) }

fig, gant = plt.subplots()

gant.set_xlim(1950, 2020)

gant.set_xlabel('years')
gant.set_ylabel('grand prixes')

for i, t in enumerate(races['name']):
    gant.broken_barh([(races['year'][i], 0.2)], (gps[t], 8), facecolor = gps_colors[t])

gant.grid(True)

gant.set_yticks([i + 5 for i in gps.values()])
gant.set_yticklabels(dict.fromkeys(races['name']).keys())
plt.show()
```

```python
drivers = pd.read_csv('../dataset/drivers.csv')
nationalities = {driver: 0 for i, driver in enumerate(dict.fromkeys(drivers['nationality'])) }

for i, n in enumerate(drivers['nationality']):
    nationalities[n] = nationalities[n] + 1

labels = []
values = []
for label, value in nationalities.iteritems():
    labels.append(label)
    values.append(value)

fig = plt.figure() 
plt.pie(values, labels = labels)
plt.show()
```

```python
ds_results = pd.read_csv('../dataset/results.csv')
ds_drivers = pd.read_csv('../dataset/drivers.csv')
ds_races = pd.read_csv('../dataset/races.csv')
ds_circuits = pd.read_csv('../dataset/circuits.csv')

driver_ref = 'hamilton'
home_country = 'UK'

m1 = pd.merge(ds_results, ds_drivers, on='driverId')
m2 = pd.merge(m1, ds_races, on='raceId')
m3 = pd.merge(m2, ds_circuits, on='circuitId')
data = m3[['driverId', 'driverRef', 'forename', 'surname', 'nationality', 'resultId', 'position', 'points', 'raceId', 'name_x', 'round', 'date', 'circuitId', 'name_y', 'location', 'country']]
data['date'] = pd.to_datetime(data['date'])
data.sort_values(by=['date'], inplace=True, ascending=True)
is_driver_data = data['driverRef'] == driver_ref
driver_data = data[is_driver_data]

is_country_driver_data = driver_data['country'] == home_country
country_driver_data = driver_data[is_country_driver_data]
total_driver_points_in_home = country_driver_data['points'].sum()

is_country_data = data['country'] == home_country
country_data = data[is_country_data]
country_data_by_driver = country_data.groupby('driverId')['points'].sum()

total_points_by_driver_in_country = country_data_by_driver.to_list()
plt.hist(x=total_points_by_driver_in_country, bins=50, facecolor="red", alpha=0.5, label="First")
plt.show()
```
